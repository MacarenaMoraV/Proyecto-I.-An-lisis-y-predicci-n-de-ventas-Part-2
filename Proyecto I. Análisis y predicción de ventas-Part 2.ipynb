#5.FILTRADO DE DATOS

° Filtrar el DataFrame para mostrar solo las filas donde las edades sean mayores a 50.

```python
# Utilicé esa columna ya que no existe la columna solicitada "Ventas"

df_filtrado_1= df[df["Age"] > 50][["Age"]]
df_filtrado_1
```

° Filtrar el DataFrame para mostrar solo las filas donde "Price per Unit" sean mayores a 40.

```python
# Utilicé esa columna ya que no existe la columna solicitada "Ventas"

df_filtrado_2= df[df["Price per Unit"] > 40][["Price per Unit"]]
df_filtrado_2
```

° Utilizando el método query(), filtrar el DataFrame para mostrar las filas donde el genero sea Femenino y las edades sean mayores a 30.

```python
# De la misma manera que procedí en los anteriores ejercicios, elegí esos parámetros para el desarrollo del ejercicio

df_filtrado_3 = df.query("Gender == 'Female' & Age > 30")[["Gender", "Age"]]
df_filtrado_3
```

#6. SLICING DE DATOS

° Selecciona y muestra solo las columnas Producto y Ventas del DataFrame.

```python
# Utilicé las columnas Transaction ID y Date, ya que no existen las columnas solicitadas
# En este caso como estoy con slicing de datos, utilizar .loc

df_slicing_1 = df.loc[:, ["Transaction ID", "Date"]]
df_slicing_1
```

° Utilizando loc[], selecciona y muestra las filas de la 5 a la 10 (inclusive) y las columnas Producto y Tienda.

```python
# Utilicé las columnas Transaction ID y Customer ID, ya que no existen las columnas solicitadas

df_slicing_2 = df.loc[5:10, ["Transaction ID", "Customer ID"]]
df_slicing_2
```
