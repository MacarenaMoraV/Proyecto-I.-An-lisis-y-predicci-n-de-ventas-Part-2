#5.FILTRADO DE DATOS

Â° Filtrar el DataFrame para mostrar solo las filas donde las edades sean mayores a 50.

```python
# UtilicÃ© esa columna ya que no existe la columna solicitada "Ventas"

df_filtrado_1= df[df["Age"] > 50][["Age"]]
df_filtrado_1
```

Â° Filtrar el DataFrame para mostrar solo las filas donde "Price per Unit" sean mayores a 40.

```python
# UtilicÃ© esa columna ya que no existe la columna solicitada "Ventas"

df_filtrado_2= df[df["Price per Unit"] > 40][["Price per Unit"]]
df_filtrado_2
```

Â° Utilizando el mÃ©todo query(), filtrar el DataFrame para mostrar las filas donde el genero sea Femenino y las edades sean mayores a 30.

```python
# De la misma manera que procedÃ­ en los anteriores ejercicios, elegÃ­ esos parÃ¡metros para el desarrollo del ejercicio

df_filtrado_3 = df.query("Gender == 'Female' & Age > 30")[["Gender", "Age"]]
df_filtrado_3
```

#6. SLICING DE DATOS

Â° Selecciona y muestra solo las columnas Producto y Ventas del DataFrame.

```python
# UtilicÃ© las columnas Transaction ID y Date, ya que no existen las columnas solicitadas
# En este caso como estoy con slicing de datos, utilizar .loc

df_slicing_1 = df.loc[:, ["Transaction ID", "Date"]]
df_slicing_1
```
